```python
import random

class Player:
    def __init__(self, name):
        self.name = name
        self.score = 0

    def roll_dice(self):
        return random.randint(1, 6)

    def toss_coin(self):
        return random.choice(['Head', 'Tail'])

    def play(self):
        dice_roll = self.roll_dice()
        if dice_roll % 2 == 0:
            self.score += 2 * dice_roll
        else:
            self.score += dice_roll

        coin_toss = self.toss_coin()
        if coin_toss == 'Head':
            self.score *= 3
        else:
            self.score -= 1

    def __str__(self):
        return f"{self.name} - {self.score}"

# Initialize players
player1 = Player("Player 1")
player2 = Player("Player 2")
player3 = Player("Player 3")
player4 = Player("Player 4")

# Game 1
print("Game 1:")
for _ in range(2):
    player1.play()
    player2.play()
print(player1)
print(player2)

# Game 2
print("\nGame 2:")
for _ in range(2):
    player3.play()
    player4.play()
print(player3)
print(player4)

# Final game
print("\nFinal game:")
if player1.score > player2.score:
    player5 = player1
else:
    player5 = player2
if player3.score > player4.score:
    player6 = player3
else:
    player6 = player4

for _ in range(2):
    player5.play()
    player6.play()
print(player5)
print(player6)

# Final winner
if player5.score > player6.score:
    print(f"\nThe winner is {player5.name}")
else:
    print(f"\nThe winner is {player6.name}")
```
